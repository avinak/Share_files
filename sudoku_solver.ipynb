### SUDOKU PUZZLE SOLVER
This script is meant to solve any sudoku puzzle thrown at it in csv format. the script iterates through the puzzle and calls different functions to validate the rules of the game and match and fix a number between 1 to 9 in each of the 81 cells of the puzzle.

First, The screenshot of the puzzle is read using tesseract engine from Google. This is then converted into a Two dimensional array. (9x9). Then laws of the games are applied to find the solution. Enjoy the code. 

First lets import necessory libraries.

```python
import pandas as pd
import numpy as np
import ast
from PIL import Image
import pytesseract
```

```python
#'f' is used for location indexing
#'tup' is the horizontal line of the sudoku puzzle
#'lst' is the vertical line of the sudoku puzzle
#'twodr' represents the two dimensinal array of the sudoku
#'number' is used to check each number from 1 to 9 and its applicability in each cell of puzzle
#'i' and 'j' are used for iterating through loop as applicable
#'l' is used to iterate through csv file


#Functions explained:
#'solution' does the actual searching and assigning of numbers and calls other functions
#'blank_locator' function checks each location if it is zero or filled with any number
#'location_checker' function calls other functions and checks all possible numbers
#'tup_checker' function checks if the number at hand in the loop is used in horizontal line
#'lst_checker' function checks if the number at hand in the loop is used in the vertical line
#'box_checker' function checks if the number at hand in the loop is used in the 3 by 3 box


def solution(twodr): 
    f=[0,0] 
    if(not blank_locator(twodr,f)): 
        return True
    tup=f[0] 
    lst=f[1] 
    for number in range(1,10): 
        if(location_checker(twodr,tup,lst,number)): 
            twodr[tup][lst]=number 
            if(solution(twodr)): 
                return True
            twodr[tup][lst] = 0     
    return False 

def blank_locator(twodr,f): 
    for tup in range(9): 
        for lst in range(9): 
            if(twodr[tup][lst]==0): 
                f[0]=tup 
                f[1]=lst 
                return True
    return False
   
        
def tup_checker(twodr,tup,number): 
    for i in range(9): 
        if(twodr[tup][i] == number): 
            return True
    return False
  

def lst_checker(twodr,lst,number): 
    for i in range(9): 
        if(twodr[i][lst] == number): 
            return True
    return False
  

def box_checker(twodr,tup,lst,number): 
    for i in range(3): 
        for j in range(3): 
            if(twodr[i+tup][j+lst] == number): 
                return True
    return False
  

def location_checker(twodr,tup,lst,number): 
    return not tup_checker(twodr,tup,number) and not lst_checker(twodr,lst,number) and not box_checker(twodr,tup - tup%3,lst - lst%3,number) 


pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
im = Image.open("screenshot1.png")

text = pytesseract.image_to_string(im, lang = 'eng')

with open("sudoku_qs.csv", 'w+') as f:
    f.writelines (text)

twodr = []


with open("sudoku_qs.csv") as quiz:
    for l in quiz:
        twodr.append(ast.literal_eval('[%s]' % l)) 
print('Screenshot scanned to find sudoku puzzle\n')
print('Please find below the unsolved 2d array of sudoku puzzle\n')
for line in twodr:
    print(line, '\n')
print("----------------------------\n")
print("Finding solutions.....\n")      
if(solution(twodr)):
    print('Please find below the 2d array of solved sudoku puzzle\n')
    for line in twodr:
        print(line, '\n') 
else: 
        print ("No solution exists")  

```
